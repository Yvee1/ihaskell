```haskell
import Diagrams.Prelude
import Diagrams.Backend.Cairo.CmdLine
```

```haskell
mkCircle :: Colour Double -> Diagram B
mkCircle color = circle 1 # fc color
                          # lw veryThick
                          # dashingG [0.2, 0.05] 0
```

```haskell
squareCircle = (mkCircle green ||| mkCircle red) === (mkCircle blue ||| mkCircle orange)
squareCircle
```

```haskell
rotateBy (1/8) squareCircle
```
