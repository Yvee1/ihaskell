Or see John Stasko, Richard Catrambone, \"An evaluation of space-filling information visualizations for depicting hierarchical structures\", 2000. <http://www.cc.gatech.edu/~john.stasko/papers/ijhcs00.pdf>.



```haskell
import Diagrams.Prelude
import Diagrams.TwoD.Sunburst
import Data.Tree (unfoldTree)

aTree = unfoldTree (\n -> (0, replicate n (n-1))) 6
example = sunburst aTree # centerXY # pad 1.1
```

```haskell
diagram example
```
