```haskell
gPart2 :: Floor -> String -> Maybe Position
gPart2 target = rec 0 0
  where rec :: Floor -> Position -> String -> Maybe Position
        rec current pos []
          | target == current    = Just pos
          | otherwise            = Nothing
        rec current pos (x:xs)
          | target == current    = Just pos
          | otherwise            = rec (current+value x) (pos+1) xs

part2 :: String -> Maybe Position
part2 = gPart2 (-1)
```
